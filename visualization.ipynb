```python
import pickle
import matplotlib.pyplot as plt 
import flwr
```

```python
# file = open("./outputs/2023-11-20/17-31-44/results.pkl", 'rb')  #0.2 
# file = open("./outputs/2023-11-20/17-38-34/results.pkl", 'rb')  #0.5
# file = open("./outputs/2023-11-20/21-38-43/results.pkl", 'rb')  #1.0  
# file = open("./outputs/2023-12-01/10-43-17/results.pkl", 'rb')# 1.0 mid 
file = open("outputs/2023-12-01/10-50-24/results.pkl", 'rb')# 1.0 end 

malicious_result = pickle.load(file)
file.close()
mr_loss = malicious_result['history'].losses_centralized
```

```python
file = open("./10rounds_clean/22-54-37/results.pkl", 'rb')
clean_result = pickle.load(file)
file.close()
cr_loss = clean_result['history'].losses_centralized
```

```python
plt.figure(figsize=(10,5))
plt.title("Loss Comparison - normal vs Malicious")
plt.plot([i for i in range(len(mr_loss))],[mr_loss[i][1] for i in range(len(mr_loss))],label="mal")
plt.plot([i for i in range(len(cr_loss))],[cr_loss[i][1] for i in range(len(cr_loss))],label="normal")
plt.xlabel("iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
# plt.figure()
mr_accuracy = malicious_result['history'].metrics_centralized['accuracy']
cl_accuracy = clean_result['history'].metrics_centralized['accuracy']
plt.title("Accuracy Comparison - normal vs Malicious")
plt.plot([i for i in range(len(mr_accuracy))],[mr_accuracy[i][1] for i in range(len(mr_accuracy))],label="mal")
plt.plot([i for i in range(len(cl_accuracy))],[cl_accuracy[i][1] for i in range(len(cl_accuracy))],label="normal")
plt.xlabel("iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
